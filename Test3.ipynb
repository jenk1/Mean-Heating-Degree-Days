```python
import pandas as pd
import numpy as np
```

```python
PopulationHeating = pd.read_table("ftp://ftp.cpc.ncep.noaa.gov/htdocs/degree_days/weighted/daily_data/1981/Population.Heating.txt", delimiter="|", skiprows=[0,1,2])
```

```python
PopulationHeating = PopulationHeating.set_index('Region')
```

```python
PopulationHeating
```

```python
# Want to go from years 1982 to 2017
i = 1982
```

```python
while(i < 1983):
    x = pd.read_table("ftp://ftp.cpc.ncep.noaa.gov/htdocs/degree_days/weighted/daily_data/%d/Population.Heating.txt" %(i), delimiter="|", skiprows=[0,1,2])
    x = x.set_index('Region')
    result = pd.concat([PopulationHeating,x], axis=1)
    i += 1
```

```python
result
```
